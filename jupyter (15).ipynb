Write a program to construct SLR and LR(0) parse table for the following grammar and
check whether the given input can be accepted or not.
Grammar:
E --> TT
T --> aT | b

Show the moves of the parser for input “aabb”, “babb”.

```python
stack = []

def reduce_stack():
    changed = False

    # T -> b
    if stack and stack[-1] == 'b':
        stack.pop()
        stack.append('T')
        print("Reduce: T -> b")
        return True

    # T -> aT
    if len(stack) >= 2 and stack[-2] == 'a' and stack[-1] == 'T':
        stack.pop()
        stack.pop()
        stack.append('T')
        print("Reduce: T -> aT")
        return True

    # E -> TT
    if len(stack) >= 2 and stack[-2] == 'T' and stack[-1] == 'T':
        stack.pop()
        stack.pop()
        stack.append('E')
        print("Reduce: E -> TT")
        return True

    return False


def parse(input_str):
    i = 0
    print("\nSTACK\tINPUT\tACTION")

    while True:
        print(''.join(stack), "\t", input_str[i:], "\t", end="")

        if i < len(input_str):
            stack.append(input_str[i])
            print("Shift", input_str[i])
            i += 1
        else:
            print("No Shift")

        while reduce_stack():
            print(''.join(stack), "\t", input_str[i:], "\tAfter Reduction")

        if ''.join(stack) == 'E' and i == len(input_str):
            print("\nString Accepted")
            return

        if i == len(input_str) and not reduce_stack():
            print("\nString Rejected")
            return


# Test cases
parse("aabb")
stack.clear()
parse("babb")
```
